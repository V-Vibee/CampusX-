**Now lets fetch the entire Data**

```python
df= pd.DataFrame()
```

```python
# Create an empty list to store the DataFrames
dfs = []

for i in range(1, 46):
    response = requests.get('https://api.themoviedb.org/3/movie/upcoming?api_key=342fff4bf91a946f37d84ea29d822846&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'release_date','title','vote_average','vote_count','original_language']]
    dfs.append(temp_df)

# Concatenate all DataFrames in the list
df = pd.concat(dfs, ignore_index=True)
```
