**Changing Date column type to DateTime**

```python
dat['date']= pd.to_datetime(dat['date'])
```

```python
dat.info()
```

**1. Extract Year**

```python
dat['date_year']= dat['date'].dt.year
dat.head()
```

**2. Extract Month**

```python
dat['date_month']= dat['date'].dt.month
dat.head()
```

```python
dat['month_name']= dat['date'].dt.month_name()
dat.head()
```

**3. Extract Day**

```python
dat['date_day']= dat['date'].dt.day
dat.head()
```

**Extract day of week**

```python
dat['week_day']= dat['date'].dt.day_of_week
dat.head()
```

```python
dat['week_name']= dat['date'].dt.day_name()
dat.head()
```

**To check if it falls on Sunday or Saturday/Weekend**

```python
dat['Weekend_Possiblity']= np.where(dat['week_name'].isin(['Sunday', 'Saturday']),1,0)
dat.head()
```

**Extract Semester**

```python
dat['date_quarter']= dat['date'].dt.quarter
dat.head()
```

```python
dat['semester']= np.where(dat['date_quarter'].isin([1,2]), 1,2)
dat.head()
```

**Time Elapsed between two dates**
