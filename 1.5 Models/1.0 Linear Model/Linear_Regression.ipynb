```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://github.com/campusx-official/100-days-of-machine-learning/raw/main/day48-simple-linear-regression/placement.csv')
```

```python
df.head()
```

**Plotting the data**

```python
plt.scatter(df['package'], df['cgpa'])
plt.xlabel('CGPA')
plt.ylabel('Package in Lacs')
plt.show()
```

```python
y= df['package']
X= df[['cgpa']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
y_pred=lr.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color='blue', label='CGPA')

# Plot the regression line
plt.plot(X_test, y_pred, color='red',label='Regression line')

plt.title('Linear Regression')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.legend()
plt.show()
```
